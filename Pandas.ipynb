# Groupby

```python
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Pandas/SalesInfo.csv")
df["Date of Purchase"]=pd.to_datetime(df["Date of Purchase"])
df["selling Ratio"] = df["Total"] / df ["Quantity"]
df["Bonus"] = df["Total"] / 10
df.head()
```

```python
#sales in everycountry -->use groupby
df["Country"].unique(), df.shape
```

```python
df.groupby("Country").sum() # use any aggregate methods
```

```python
df.groupby(["Country","Region"]).sum() # use any aggregate methods
```

```python
df.groupby(["Country","Region", "Sales Person"]).sum()
```

```python
df["Year"] = df["Date of Purchase"].dt.year
df.groupby(["Country", "Region", "Year"])["Quantity"].sum()
```

```python
df.groupby(["Country", "Region",df["Date of Purchase"].dt.year])["Quantity"].sum()
```

```python
df.groupby(["Country", "Region",df["Date of Purchase"].dt.year])["Quantity"].agg(["sum", "min", "max", "mean"])
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Pandas/SalesInfo.csv")
df["Date of Purchase"]=pd.to_datetime(df["Date of Purchase"])
df["selling Ratio"] = df["Total"] / df ["Quantity"]
df["Bonus"] = df["Total"] / 10
```

```python
new_df = pd.DataFrame()
new_df["Employee Name"] = df["Sales Person"].unique()
import random
exp = [1,2,4]
random.shuffle(exp)
new_df["Experience in Company"] =exp
new_df["Total Experience"] = new_df["Experience in Company"] + random.randint(1,10)
new_df
```

```python
df.merge(new_df, left_on="Sales Person", right_on="Employee Name").drop("Employee Name", axis=1)
```

```python
df.join(new_df).head() # look at employee name , it has merged only the indexes
```

```python
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/Pandas/SalesInfo.csv")

new_df = pd.DataFrame()
new_df["Employee Name"] = df["Sales Person"].unique()
import random
exp = [1,2,4]
random.shuffle(exp)
new_df["Experience in Company"] =exp
new_df["Total Experience"] = new_df["Experience in Company"] + random.randint(1,10)

df.set_index("Sales Person", inplace=True)
new_df.set_index("Employee Name", inplace=True)
df.join(new_df).head()
```
